### Congruence Relation
### $x_{i+1} = (ax_{i}+b)$ $mod$ $m$
### $u_{i+1} = x_{i+1}/m$

```python
import numpy as np
import matplotlib.pyplot as plt
```

## Q.1

```python
# Function to generate sequence of numbers using General Linear Congruence Relation
def generate_sequence(a, b, m, x0, n):
    x = np.zeros(n + 1, dtype=int)
    x[0] = x0
    for i in range(1, n + 1):
        x[i] = (a * x[i - 1] + b) % m
    return x
```

```python
# Parameters : a = 6, b = 0, m = 11
sequences = [generate_sequence(6, 0, 11, x0, 10) for x0 in range(10)]
for i in range(0, 10):
    print("x0 =",i,":",sequences[i])
print("")
# Parameters : a = 3, b = 0, m = 11
sequences = [generate_sequence(3, 0, 11, x0, 10) for x0 in range(10)]
for i in range(0, 10):
    print("x0 =",i,":",sequences[i])
```

### Q.2

```python
# Function to generate normalized sequence using General Linear Congruence Relation
def generate_normalized_sequence(a, m, x0, n):
    x = np.zeros(n, dtype=int)
    u = np.zeros(n)
    x[0] = x0
    for i in range(1, n):
        x[i] = (a * int(x[i - 1])) % m
        u[i] = x[i] / m
    return u
```

```python
# Taking Parameters given in question and showing the plots.
m = 244944
a = [1597, 51749]
x0_values = [1, 2, 3, 4, 5]
group_ranges = np.arange(0, 1.05, 0.05)

for x0 in x0_values:
    u = generate_normalized_sequence(a[0], m, x0, 10000)
    frequencies, _ = np.histogram(u, bins=group_ranges)
    plt.bar(group_ranges[:-1], frequencies, width=0.05, align='edge')
    plt.title(f'Frequency Distribution for x0 = {x0}, m = {m}, a = {a[0]}')
    plt.xlabel('Range')
    plt.ylabel('Frequency')
    plt.show()

for x0 in x0_values:
    u = generate_normalized_sequence(a[1], m, x0, 10000)
    frequencies, _ = np.histogram(u, bins=group_ranges)
    plt.bar(group_ranges[:-1], frequencies, width=0.05, align='edge')
    plt.title(f'Frequency Distribution for x0 = {x0}, m = {m}, a = {a[1]}')
    plt.xlabel('Range')
    plt.ylabel('Frequency')
    plt.show()
```

### Q.4

```python
def linear_congruential_generator(a, b, m, x0, n=17):
    x = np.zeros(n, dtype=int)
    x[0] = x0
    for i in range(1, n):
        x[i] = (a * x[i - 1] + b) % m
    return x

def recursion(U, N):
    for i in range(17, N):
        Ui = U[i - 17] - U[i - 5]
        if Ui < 0:
            Ui += 1
        U = np.append(U, Ui)
    return U
```

```python
# Initial sequence using linear congruence generator
m = 2048
a = 1229
b = 1
x0 = 12345
U_init = linear_congruential_generator(a, b, m, x0)

# Generate values for N = 1000, 10000, 100000
for N in [1000, 10000, 100000]:
    U = recursion(U_init, N)
    plt.hist(U, bins=50)
    plt.title(f'Histogram of U values for N = {N}')
    plt.xlabel('Value')
    plt.ylabel('Frequency')
    plt.show()

    plt.scatter(U[:-1], U[1:], s=1)
    plt.title(f'Scatter Plot of U(i) vs U(i+1) for N = {N}')
    plt.xlabel('U(i)')
    plt.ylabel('U(i+1)')
    plt.show()
```
